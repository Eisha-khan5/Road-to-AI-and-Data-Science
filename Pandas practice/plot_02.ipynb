# Plotting in python
## Using iris dataset

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

flower = sns.load_dataset('iris')
flower
```

```python
flower.columns
```

```python
flower.info()
```

```python
flower.isnull().sum()
```

```python
#bar plot
flower.petal_length.value_counts().plot(kind= 'bar')
```

```python
sns.countplot(x= 'petal_width' , hue='species', data=flower)
```

```python
plt.scatter(data=flower ,x='petal_width' , y='petal_length', c='seagreen')
```

```python
sns.scatterplot(data=flower ,x='petal_width' , y='petal_length', hue='species')
```

```python
sns.barplot(data=flower ,x='petal_width' , y='petal_length', hue='species')
```

```python
sns.pairplot(flower)
```

```python
flower.hist(grid=False, figsize=(9,7))
```

```python
sns.boxenplot(data=flower ,x='petal_width' , y='petal_length', hue='species')
```

```python
# palettte ='deep' , 'muted', 'bright' ,'pastel' , 'dark' , 'colorblind' , 'viridis', 'plasma' , 'inferno', 'cividis', 'magma' 
g= sns.FacetGrid(flower , hue='species' , col='species', margin_titles=True,
                 palette='magma')
g= g.map(plt.scatter, 'petal_width','petal_length' , edgecolor= 'w').add_legend()
```

```python
sns.stripplot(data=flower ,x='petal_width' , y='petal_length', hue='species')
```

```python
numeric_data = flower.select_dtypes(include=[np.number])
corr = numeric_data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title('Heat map of iris dataset')
```

```python
sns.violinplot(data=flower ,x='petal_width' , y='petal_length', hue='species')
```
