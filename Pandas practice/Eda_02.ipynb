# EDA (Exploratory Data Analysis)
Exploratory data analysis is a method used to `analyze` and `summarize` datasets. In python, you can use pandas library to perform EDA on a dataset. 

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#analyze
df = sns.load_dataset('titanic')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#summarize (summary statistics)
df.describe()
```

```python
# find null values
df.isnull().sum()
```

```python
# to find percentage
df.isnull().sum() * 100 / len(df)
```

```python
# finding no of unique values
df.nunique()
```

```python
df.columns
```

```python
#visualize the nan values in heatmap
sns.heatmap(df.isnull())
```

```python
sns.heatmap(df.isnull(), yticklabels=False , cbar= False, cmap='viridis')
```

```python
#drop deck
df.drop(columns='deck' , inplace=True)
```

```python
df.shape
```

```python
# visualize and explore
# in a count plot y axis will represnt the count , so we provide attribute of x axis only
sns.countplot(data= df, x='survived')
```

```python
sns.barplot(data= df, x='survived' , y='age', hue='who')
```

```python
#data distribution
sns.histplot(data= df , x='age', bins=30)
```

```python
# aplha is opacity (brightness)
df["age"].hist(bins = 30, color= 'darkred' , alpha=0.75)
```

```python
#whenever we make a boxplot , on one side we take catagorical variable and on the other side numerical variable
sns.boxplot(data=df, y='age', x='survived', hue='who')
```

```python
#scatterplot has numeric values on both axis
#apply limit to remove outliers
sns.scatterplot(x='age', y='fare', data=df, hue='pclass')
plt.ylim(0,300)
```

```python
sns.pairplot(df, hue='survived')
```

```python
#stacked barchart/plot
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
#swarmplot
sns.swarmplot(data=df, x='survived' , y='age' , hue='who')
```

```python
sns.catplot(x='pclass' , y='fare', hue='embarked' , kind='point', data=df)
```

```python
# to check relation of two numeric variables we use lmplot
sns.lmplot(data=df , y='fare' , x='age')
```

```python
#remove outliers , for that apply condition
sns.lmplot(data=df[df.fare<300], x='age', y='fare', hue='pclass')
```

```python
#how to add column and make a group
sns.histplot(data=df, x='age')
```

```python
#grouping
df.columns
```

```python
#to add a new column
df['age_groups'] = pd.cut(df['age'], [0, 10, 18, 30, 50, 80 ], 
                labels=['Kids' , 'Child', 'Youth', 'Middle Aged', 'Old'])
df.head()
```

```python
df.groupby("age_groups")['survived'].mean().plot(kind='bar')
```

```python
#inside varse , add those variable names you want to include in pairplot
sns.pairplot(df, hue='survived', vars= ['fare', 'age', 'pclass'])
```
