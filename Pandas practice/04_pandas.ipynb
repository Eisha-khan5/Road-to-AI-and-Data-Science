# Data selection / Subsetting , Slicing

```python
# (,) separates the values of rows and columns
# when we want to specify on the basis of index then we use iloc
df.iloc[0:2 , 0:3] 
```

```python
#only colon means all the rows
df.iloc[: , 0:3] 
```

```python
# to mention the column name in string we use loc 
df.loc[:, ["A","B"]]
```

```python
#gives the values at third index
df.iloc[3]
```

```python
df.iloc[3:5, 0:2]
```
