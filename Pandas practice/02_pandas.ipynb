# 16- Splitting a string into multiple columns 

```python
# 
import pandas as pd
df = pd.DataFrame({'Name': ['Samman Fatima' , 'Tayyaba sajid' , 'Eisha Khan' , 'Bismah Akhlaq'] ,
                   'location': ['Mianwali, Pakistan' , 'HassanAbdal , Pakistan' , 'OfficersColony , WahCantt' , 'Abbotabad,Pakistan']})
df 
```

```python
# now convert the name column into first and last names , similarly location into city and country
#1. Splitting the column into two columns
df.Name.str.split(' ' , expand = True)
```

```python
#2.Adding those splits into new columns
df[["first_name" , "last_name"]] = df.Name.str.split(' ' , expand = True)
df
```

```python
# splitting the location column
df[["City" , "Country"]] = df.location.str.split(',' , expand = True)
df
```

```python
# redefine the dataset columns , include the column names that we want in our dataset
# refine data manipulation
df = df[['first_name' , 'last_name' , 'City' , 'Country']]
df
```
