# 1. Import dataset 

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import dataset
df = sns.load_dataset('titanic')
df
```

# 2. Bigger picture of data

```python
#check the datatypes
df.dtypes
```

```python
#find missing or null values
df.info()
```

```python
df.isnull().sum()
```

```python
# df will remain original dataset , new updated values of age will be stored in df2
df2 = df
```

```python
#find the percentage of null values
# general rule (if a column has 70% or 75% null values then that column becomes meaningless , we may delete it)
# and for the columns having less than 1% we delete the specific rows containing null values (e-g : embarked, embark_town)
# for the column of age (less than 30 , 40%) we will impute it using 3, 4 method 
# 1. replace the missing values with the avg age 
# 2. replace the values with mean , median or mode 

# catagorical = replace missing val with most frequent value (mode) , forward/backward fill
# numeric = Replace missing val with the mean or median of the column. 
# boolean = Mode Imputation: Replace missing val with the most frequent value (T or F), 
# Convert boolean data to 0 and 1 and then use techniques similar to numeric data imputation.

df2.isnull().sum() *100 / len(df2)
```

```python
df2['age'].unique()
```

```python
print(df2['age'].max())
print(df2['age'].min())
print(df2['age'].mean())
```

```python
# fill the null values but dont alter the original dataset
df2['age'].fillna(df2['age'].mean(), inplace=True)
```

```python
df2
```

```python
# df2 will remain same , new updated values of fare will be stored in df3
df3 = df2
```

```python
df3.isnull().sum()
```

```python
df3 = df3.drop(columns='deck')
```

```python
df3.isnull().sum()
```

```python
#df3 will remain same , new updated values of embarked will be stored in df4
df4 = df3
```

```python
# replace missing values with first mode
df4['embark_town'].fillna(df4['embark_town'].mode()[0] , inplace=True )
df4['embarked'].fillna(df4['embarked'].mode()[0] , inplace=True )
```

```python
df4.isnull().sum()
```

```python
# catagorical variables dont have mean , they have occurence
print('mean = ', df['age'].mean())
print('median = ', df['age'].median())
print('mode = ', df['embark_town'].mode())
```
