# Plots using Pandas

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
g= sns.FacetGrid(df , hue='who' , col='who', margin_titles=True,
                 palette='magma')
g= g.map(plt.scatter, 'age','fare' , edgecolor= 'w').add_legend()
```

```python
# stacked bar chart using pandas
df.groupby('survived')[['age', 'fare']].mean().plot.bar(stacked=True)
```

```python
# whole bar chart using pandas
df.groupby(['survived', 'class'])[['age', 'fare']].mean().plot.bar()
```

```python
# draw a pie chart that shows % of who survived
df.groupby('who')['survived'].mean().plot.pie()
```

```python
# draw map using pandas that shows embark_town
df.groupby('embark_town')['survived'].mean().plot.bar(color='#731f54')
```

```python
df.describe()
```

```python
# draw a scatter plot using pandas
sns.scatterplot(data=df, y='fare', x='age', hue='pclass')
```

```python
df.plot.scatter(x="age", y="fare", c="class", cmap="viridis")
```

```python
# Hexagonal Bin plot
df.plot.hexbin(x="age", y="fare", gridsize=25, C="pclass")
```
