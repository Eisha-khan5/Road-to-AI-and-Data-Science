# Seaborn and Plotting

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data 
df = sns.load_dataset('titanic')
```

```python
df.info()
```

## Scatter Plot

```python
# Scatter  plot
# scatter plot size or figure size
plt.figure(figsize=(8,5))
sns.scatterplot(x = 'age', y = 'fare', data = df, hue='sex')
```

```python
# Scatter  plot
# scatter plot size or figure size
plt.figure(figsize=(7,5))
# change the scatter point colors    "#e74c3c","#34495e", "#3498db"
pallete = {'male': "#9b59b6",'female': "#2ecc71"}
sns.scatterplot(x = 'age', y = 'fare', data = df, hue='sex', palette=pallete)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of age vs fare')
```

```python
# Scatter  plot
# scatter plot size or figure size
plt.figure(figsize=(7,5))
# change the scatter point colors    "#e74c3c","#34495e", "#3498db"
pallete = {'male': "#9b59b6",'female': "#2ecc71"}
sns.scatterplot(x = 'age', y = 'fare', data = df, hue='sex', palette=pallete)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of age vs fare')
# Saving a plot 300dpi (dots per inch)
plt.savefig('./Plots/Scatter_plot_age_vs_fare.png', dpi=300)
# Saving a plot in vector format svg
plt.savefig('./Plots/Scatter_plot_age_vs_fare.svg')
```

```python
# style of a plot or the theme of a plot
sns.set_theme(style='darkgrid',font_scale=1,font='Comic Sans MS')
plt.figure(figsize=(7,5))
# change the scatter point colors    "#e74c3c","#34495e", "#3498db"
pallete = {'male': "#9b59b6",'female': "#2ecc71"}
sns.scatterplot(x = 'age', y = 'fare', data = df, hue='sex', palette=pallete)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of age vs fare')
```

Interpretation:
1. We have Age of passengers at x-axis and Fare at y-axis
2. The scatter plot shows the relationship between age and fare of titanic passangers.
3. We further grouped the passangers by their genders such as male (Purple points) and female (Green points).
4. Two male passangers gave the highest fare in titanic, and they might be travelling in the first class.
5. The oldest person at titanic is also a male having 80 years of age as shown in the scatter plot.(As shown in Fig 1)

## Box Plot

```python
# style of a plot or the theme of a plot
sns.set_theme(style='darkgrid',font_scale=1,font='Comic Sans MS')
plt.figure(figsize=(12,6))
# change the scatter point colors    "#e74c3c","#34495e", "#3498db"
pallete = {'First': "#9b59b6",'Second': "#2ecc71",'Third': "#3498db"}
sns.boxplot(x = 'sex', y = 'age', data = df, hue='class', palette=pallete)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Box plot of age vs fare')
# change the legend position and title
plt.legend(loc = 'upper right', title = 'Class Name')
```

## Bubble Plot

```python
# style of a plot or the theme of a plot
sns.set_theme(style='darkgrid',font_scale=1,font='Comic Sans MS')
plt.figure(figsize=(10,5))
# change the scatter point colors    "#e74c3c","#34495e", "#3498db"
pallete = {'male': "#9b59b6",'female': "#2ecc71"}
sns.scatterplot(x = 'age', y = 'fare', data = df, hue='sex',size='age', palette=pallete)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of age vs fare')
```

## Boxen Plot

```python
# style of a plot or the theme of a plot
sns.set_theme(style='darkgrid',font_scale=1,font='Comic Sans MS')
plt.figure(figsize=(12,6))
# change the scatter point colors    "#3498db","#9b59b6" ,blue="#2ecc71"
pallete = {'First': "#e74c3c",'Second':"#34495e" ,'Third': "#3498db"}
sns.boxenplot(x = 'sex', y = 'age', data = df, hue='class', palette=pallete)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Boxen plot of age vs fare')
# change the legend position and title
plt.legend(loc = 'upper right', title = 'Class Name')

# change the x ticks names
plt.xticks(ticks=[0,1],labels=['Male','Female'], rotation=45)
```

```python
# make scatter plot with plotly
import plotly.express as px
fig = px.scatter(df, x="age", y="fare", color="sex")
fig.show()
```

##  Bar Plot

```python
# style of a plot or the theme of a plot
sns.set_theme(style='darkgrid',font_scale=1,font='Comic Sans MS')
plt.figure(figsize=(12,6))
# change the scatter point colors    "#3498db","#9b59b6" ,blue="#2ecc71"
pallete = {'First': "#e74c3c",'Second':"#34495e" ,'Third': "#3498db"}
sns.barplot(x = 'sex', y = 'age', data = df, hue='class', palette=pallete, capsize=0.1)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Bar plot of age vs fare')
# change the legend position and title
plt.legend(loc = 'upper right', title = 'Class Name')

# change the x ticks names
plt.xticks(ticks=[0,1],labels=['Male','Female'], rotation=45)
```

## Violin Plot

```python
# style of a plot or the theme of a plot
sns.set_theme(style='darkgrid',font_scale=1,font='Comic Sans MS')
plt.figure(figsize=(12,6))
# change the scatter point colors    "#3498db","#9b59b6" ,blue="#2ecc71"
pallete = {'First': "#e74c3c",'Second':"#34495e" ,'Third': "#3498db"}
sns.violinplot(x = 'sex', y = 'age', data = df, hue='class', palette=pallete)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Bar plot of age vs fare')
# change the legend position and title
plt.legend(loc = 'upper right', title = 'Class Name')

# change the x ticks names
plt.xticks(ticks=[0,1],labels=['Male','Female'], rotation=45)
```
